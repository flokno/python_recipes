```python
import pandas as pd
import seaborn as sns
```

```python
# see https://github.com/mwaskom/seaborn-data for list of datasets

df = sns.load_dataset('planets')
df.head()
```

```python
# Store to hdf5

name = "DataFrame"
file = "test.h5"

with pd.HDFStore(file) as store:
    store[name] = df
```

```python
# read from hdf5
with pd.HDFStore(file) as store:
    print("Stored objects:")
    for k in store:
        print(f"  {k}")

    df = store[name]

df.head()
```
